# Scatter plot antara AVG. Area Income dan Price

```python

plt.scatter(data['Avg. Area Income'],data['Price'])
plt.xlabel('Avg. Area Income')
plt.ylabel('Price')
plt.show()
```

Berdasarkan scatter plot diatas, terlihat ada data outlier. 
Sehingga harus dihilangkan terlebih dahulu

```python
import numpy as np
area = data['Avg. Area Income']
harga = data['Price']


meanArea = np.mean(area)
meanHarga = np.mean(harga)
stdArea = np.std(area)
stdHarga = np.std(harga)

dataClean = data[(np.abs((data['Avg. Area Income']-meanArea)/stdArea) <= 3) & (np.abs((data['Price']-meanHarga)/stdHarga) <= 3)]
```

```python
import matplotlib.pyplot as plt
plt.scatter(dataClean['Avg. Area Income'], dataClean['Price'])
plt.show()
```

Diperoleh Scatter Plot baru antara Avg. Area Income dan Price setelah menghilangkan outlier.

# Regresi Linear

y = a + b * x

y = variabel dependen (Price)

x = variabel independen (Avg. Area Income)

a = intersept
 
b = coef. Avg. Area Income

```python
dataClean
```

```python
dataClean.info()
```

```python
data2=dataClean.filter(items=['Avg. Area Income','Price'])
data2.head()
```

```python
x=data2[['Avg. Area Income']]
y=data2['Price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=101) 
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression() 
lm.fit(x_train,y_train)
```

```python
print('intercept:',lm.intercept_)
print('slope:',lm.coef_)
```

```python
cdf=pd.DataFrame(lm.coef_,x.columns,columns=['Coeff']) 
cdf
```

Diperoleh bentuk persamaan regresi linearnya adalah

#  y = -226136.21 + 21.23 * x

# Prediction test data

```python
predictions=lm.predict(x_test)
predictions
```

```python
plt.scatter(y_test,predictions)
plt.show()
```


 # Evaluasi Model
Mean Absolute Error (MAE) is the mean of the absolute value of the errors:

 $\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$
 
Mean Squared Error (MSE) is the mean of the squared errors:

$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$

Root Mean Squared Error (RMSE) is the square root of the mean of the squared errors:

$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$


```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn import metrics

print('Mean absolute error: %.2f' % np.mean(np.absolute(predictions-y_test)))
print('MSE: %.2f' % mean_squared_error(y_test, predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test,predictions)))
from sklearn.metrics import r2_score
print('R2-score: %.2f' % r2_score(y_test, predictions))
```

# Residu

```python
sns.distplot((y_test-predictions),bins=50);
```

```python
def mean_absolute_percentage_error(y_test, predictions): 
    y_test, y_pred = np.array(y_test), np.array(predictions)
    return np.mean(np.abs((y_test - predictions) / y_test)) * 100
mean_absolute_percentage_error(y_test, predictions)
```

# K Means Clustering

K-means Clustering. Menggunakan kata ‘K’ artinya kita akan membentuk banyak klaster sebanyak K kelompok. K-means ini merupakan metode yang paling sederhana dan banyak digunakan untuk memisahkan dataset kedalam ‘k’ kelompok.

```python
conda install -c districtdatalabs yellowbrick
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
## Menggunakan metode elbow untuk menentukan angka cluster yang tepat
model = KMeans(random_state=0)
visualizer = KElbowVisualizer(model, k=(2,10), metric='silhouette', timings=False)
```

```python
#fit data dan visualisasi
visualizer.fit(P)    
visualizer.poof()
```

Terdapat banyak cara dalam menentukan jumlah cluster terbaik, salah satunya dengan menggunakan metode Silhouette. Berdasarkan outputan data diperoleh k = 7, berarti jumlah cluster ada 7 macam.



```python
#Applying kmeans to the dataset / Creating the kmeans classifier
kmeans = KMeans(n_clusters = 7, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(P)
```

```python
#Visualising the clusters
plt.scatter(P[y_kmeans == 0, 0], P[y_kmeans == 0, 1],   s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(P[y_kmeans == 1, 0], P[y_kmeans == 1, 1],   s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(P[y_kmeans == 2, 0], P[y_kmeans == 2, 1],   s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(P[y_kmeans == 3, 0], P[y_kmeans == 3, 1],   s = 100, c = 'pink', label = 'Cluster 4')
plt.scatter(P[y_kmeans == 4, 0], P[y_kmeans == 4, 1],   s = 100, c = 'black', label = 'Cluster 5')
plt.scatter(P[y_kmeans == 5, 0], P[y_kmeans == 5, 1],   s = 100, c = 'purple', label = 'Cluster 6')
plt.scatter(P[y_kmeans == 6, 0], P[y_kmeans == 6, 1],   s = 100, c = 'brown', label = 'Cluster 7')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'yellow', label = 'Centroids')

plt.legend()
```
